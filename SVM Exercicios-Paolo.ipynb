1) Explique o que é conjuntos de **treino** e conjuntos de **teste** e porque esses conjuntos recebem esse nome.

Resposta: Conjunto de treino é a porção do dataframe separada para treinar o algorítmo, para que ele aprenda com os dados a fazer a predição dos dados.

Conjunto de teste é a porção dos dados que será usada para testar o modelo de predição do algoritmo, separada do conjunto de treino e o algorítmo não conhece esses dados. 

Depois o objetivo é compararmos como o algorítmo se saiu com os dados de teste a partir do que ele aprendeu no treino e usar a matriz de confusão para avaliar os erros.

13) Crie um pipeline chamado **MD_01** com as seguintes configurações:

    a) Standardscaler como a primeira etapa.
    b) Classificador SVM com Kernel RBF,  C=100 , gamma= 0.01.

Edite as células abaixo.

```python
# Importe as bibliotecas de Pipelines e Pré-processadores

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Criando Pipeline

MD_01 = Pipeline([
    ('scaler', StandardScaler()),
    ('classificador_svm', svm.SVC(C=100, gamma=0.01))
])
```

```python
MD_01.steps
```

14) Teste o desempenho do seu modelo (pipeline) com a função acurácia

```python
# Função que retorna a acurácia após fazer um validação cruzada (cross validation)
def Acuracia(classificador_svm,X,y): #classificador, x dados de treino, y dados de classe
    resultados = cross_val_predict(classificador_svm, X, y, cv=10) #cross_val_predict é o método que faz a validação cruzada do sklearn, com dados de treino, com as classes, cv é o número de partes que vamos separar os dados
    return metrics.accuracy_score(y,resultados) #aqui estamos criando a acruácia de acordo com os valores reais (y) e os valores encontrados (resultados)
```

```python
# Teste o modelo usando o pipeline(criado anteriormente)
Acuracia(MD_01,df_spotify,classes)
```

```python
# Importa o utilitário GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# Lista de Valores de C
lista_C = [0.001, 0.01, 0.1, 1, 10, 100]

# Lista de Valores de gamma
lista_gamma = [0.001, 0.01, 0.1, 1, 10, 100]

# lista_kernel = ['rbf', 'poly'] #'linear'

#objetivo da lista é validar o modelo com vários valores de C e gamma para ver o resultado do algoritmo com cada um deles.
```

```python
# Define um dicionário que recebe as listas de parâmetros e valores.
parametros_grid = dict(classificador_svm__C=lista_C, classificador_svm__gamma=lista_gamma) #, classificador_svm__kernel=lista_kernel)



# o classificador vai receber no objeto C a lista C e o classificador vai receber no parâmetro gamma a lista gamma
```

```python
# Objeto Grid recebe parâmetros de Pipeline, e configurações de cross validation
grid = GridSearchCV(MD_01, parametros_grid, cv=10, scoring='accuracy')

# o grid search é integrado ao svm, os pipelines são objetos que encapsulam pré-processamento e treino do algoritmo.
# Estamos passando no grid search, o pip3 que tem o standard scaller e o classificador com kernel rbf
# Estamos passando também o parametros_grid que tem o C e Gamma com os valores definidos na lista acima. CV=10 é o número de folds. Scoring = 'accuracy' é a métrica de avaliação
# poderiamos usar outra métrica como regressão (erro médio quadrático) etc. Usamos a acurácia pois é a utilizada nas aulas anteriores.
```

```python
grid
```

```python
# Aplica o gridsearch passando os dados de treino e classes.
grid.fit(df_spotify,classes)

#passamos todos os dados e todas as classes

#Agora estamos efetivamente aplicando o grid search, vai fazer essa série de validações e trazer os resultados.
#Estamos usando o método fit que vai aplicar o pipeline, vai aplicar os passos de scaler, aplicar o algorítmo de svm, vai variar todos os parâmetros da lista c e gamma,
# vai aplicar c 0.001 e gamma 0.001, 0.01 e 0.01 e assim por diante...
```

```python
# Imprime os scores por combinações
grid.cv_results_

#Este comando imprime todos os scores por configurações.
```

```python
# Imprime os melhores parâmetros
grid.best_params_
```
